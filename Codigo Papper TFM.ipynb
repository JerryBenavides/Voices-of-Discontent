```python
# Columnas por bloque temático
Tema01_Discurso_contra_elites_politicas_vars = [
    'la casta', 'los de arriba', 'paguitas', 'chiringuitos', 'corruptos',
    'Soros', 'elites globalistas',
]

Tema02_Discurso_contra_elites_economicas_vars = [
    'IBEX 35', 'poderes económicos',
    'poderes oligárquicos', 'grandes fortunas', 'elite económica',
    'neoliberalismo', 'grandes tecnológicas', 'empresas energéticas'
 ]

Tema03_Discurso_contra_élites_mediáticas_vars = [
    'poderes mediáticos', 'izquierda mediática', 'medios subvencionados',
    'medios del bipartidismo'
]

Tema04_Discurso_contra_élites_culturales_Limpio_vars = [
    'lobby LGTB', 'chiringuitos de género',
    'ideología de género', 'propaganda progre', 'dictadura climática'
]

Tema05_Representación_positiva_del_pueblo_vars = [
    'pueblo', 'patria'
]

Tema06_Pueblo_soberano_vars = [
    'soberanía española', 'soberanía nacional '
]

Tema07_Pueblo_como_clase_social_vars = [
    'los de abajo', 'españoles de a pie', 'España que madruga', 'trabajadores del campo', 'España real', 'gente trabajadora'
]

Tema08_Pueblo_como_nación_vars = [
    'la España viva', 'nuestro pueblo', 'la nación'
]

Tema09_Defensa_de_la_voluntad_popular_vars = [
    'voluntad de los españoles', 'voluntad del pueblo', 'voluntad nacional '
]

Tema10_Enemigo_de_izquierdas_vars = [
    'fascismo', 'extrema derecha', 'ultraderecha', 'franquistas', 'negacionistas'
]

Tema11_Enemigo_de_derecha_vars = [
    'separatistas', 'comunistas', 'proetarra', 'chavismo', 'Venezuela',
    'partidos golpistas', 'progres', 'menas', 'okupas', 'magrebíes',
    'separatismo totalitario', 'república islámica', 'ley trans', 'inmigración ilegal'
]
```

```python
# Diccionario de temas
temas_dict = {
    'Tema 01': Tema01_Discurso_contra_elites_politicas_vars,
    'Tema 02': Tema02_Discurso_contra_elites_economicas_vars,
    'Tema 03': Tema03_Discurso_contra_élites_mediáticas_vars,
    'Tema 04': Tema04_Discurso_contra_élites_culturales_Limpio_vars,
    'Tema 05': Tema05_Representación_positiva_del_pueblo_vars,
    'Tema 06': Tema06_Pueblo_soberano_vars,
    'Tema 07': Tema07_Pueblo_como_clase_social_vars,
    'Tema 08': Tema08_Pueblo_como_nación_vars,
    'Tema 09': Tema09_Defensa_de_la_voluntad_popular_vars,
    'Tema 10': Tema10_Enemigo_de_izquierdas_vars,
    'Tema 11': Tema11_Enemigo_de_derecha_vars,
}

# Contar tweets que activan cada tema ( ≥ 0.8)
tema_resultados = {}
for tema, variables in temas_dict.items():
    count = df_fuerte_detalle[variables].ge(0.8).any(axis=1).sum()
    tema_resultados[tema] = count

# Mostrar
import pandas as pd
df_temas_activos = pd.DataFrame.from_dict(tema_resultados, orient='index', columns=['Tweets fuertes (≥ 0.8)'])
print(df_temas_activos)
```

```python
# Crear las variables promedio por tema 
df_fuerte_detalle = df_fuerte_detalle.copy()  

df_fuerte_detalle['tema_01'] = df_fuerte_detalle[Tema01_Discurso_contra_elites_politicas_vars].mean(axis=1)
df_fuerte_detalle['tema_02'] = df_fuerte_detalle[Tema02_Discurso_contra_elites_economicas_vars].mean(axis=1)
df_fuerte_detalle['tema_03'] = df_fuerte_detalle[Tema03_Discurso_contra_élites_mediáticas_vars].mean(axis=1)
df_fuerte_detalle['tema_04'] = df_fuerte_detalle[Tema04_Discurso_contra_élites_culturales_Limpio_vars].mean(axis=1)
df_fuerte_detalle['tema_05'] = df_fuerte_detalle[Tema05_Representación_positiva_del_pueblo_vars].mean(axis=1)
df_fuerte_detalle['tema_06'] = df_fuerte_detalle[Tema06_Pueblo_soberano_vars].mean(axis=1)
df_fuerte_detalle['tema_07'] = df_fuerte_detalle[Tema07_Pueblo_como_clase_social_vars].mean(axis=1)
df_fuerte_detalle['tema_08'] = df_fuerte_detalle[Tema08_Pueblo_como_nación_vars].mean(axis=1)
df_fuerte_detalle['tema_09'] = df_fuerte_detalle[Tema09_Defensa_de_la_voluntad_popular_vars].mean(axis=1)
df_fuerte_detalle['tema_10'] = df_fuerte_detalle[Tema10_Enemigo_de_izquierdas_vars].mean(axis=1)
df_fuerte_detalle['tema_11'] = df_fuerte_detalle[Tema11_Enemigo_de_derecha_vars].mean(axis=1)

# Agrupar por autor y calcular el promedio por tema
promedio_por_tema_y_autor = df_fuerte_detalle.groupby("Author")[[
    'tema_01', 'tema_02', 'tema_03', 'tema_04', 'tema_05',
    'tema_06', 'tema_07', 'tema_08', 'tema_09', 'tema_10', 'tema_11'
]].mean().round(2)  

# Mostrar resultado
print(promedio_por_tema_y_autor)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Diccionario de agrupación (ya aplicado previamente)
agrupacion_partidos = {
    'ppopular': 'PP',
    'dolorsmm': 'PP',
    'psoe': 'PSOE',
    'teresaribera': 'PSOE',
    'sumar': 'Sumar',
    'estrella_galan': 'Sumar',
    'podemos': 'Podemos',
    'irenemontero': 'Podemos',
    'vox_es': 'Vox',
    'jorgebuxade': 'Vox'
}

# 2. Crear columna de partido
df_fuerte_detalle['Partido'] = df_fuerte_detalle['Author'].map(agrupacion_partidos)

# 3. Agrupar por partido y calcular promedios
bloques = ['antielitista', 'pueblo_uno', 'voluntad_pueblo', 'enemigo_izquierda', 'enemigo_derecha']

promedio_bloques_por_partido = df_fuerte_detalle.groupby('Partido')[bloques].mean().round(2)

# 4. Traducir los nombres de los bloques al inglés
promedio_bloques_por_partido = promedio_bloques_por_partido.rename(columns={
    'antielitista': 'Anti-elite discourse',
    'pueblo_uno': 'People',
    'voluntad_pueblo': 'Popular will',
    'enemigo_izquierda': 'Left-wing adversaries',
    'enemigo_derecha': 'Right-wing adversaries'
})

# 5. Orden específico de partidos
orden_partidos = ['Podemos', 'Sumar', 'PSOE', 'PP', 'Vox']

# 🔵 PRIMER GRÁFICO: Heatmap de promedios (valores numéricos)
plt.figure(figsize=(10,6))
sns.heatmap(
    promedio_bloques_por_partido.loc[orden_partidos],
    annot=True, fmt=".2f", cmap="YlOrRd", linewidths=0.5, linecolor='gray'
)
plt.title("Heatmap of Average Thematic Blocks by Party")
plt.xlabel("Thematic blocks")
plt.ylabel("Political party")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("heatmap_promedios_bloques.png", dpi=300)
plt.show()

# 🔵 SEGUNDO GRÁFICO: Heatmap categorizado en percentiles
# Convertir a percentiles
percentiles = promedio_bloques_por_partido.rank(pct=True)

# Función para categorizar
def categorizar(valor):
    if valor <= 0.25:
        return 'Very Low'
    elif valor <= 0.50:
        return 'Low'
    elif valor <= 0.75:
        return 'Medium'
    else:
        return 'High'

# Aplicar categorías
percentiles_categoricos = percentiles.applymap(categorizar)

# Graficar
plt.figure(figsize=(10,6))
sns.heatmap(
    percentiles.loc[orden_partidos],
    annot=percentiles_categoricos.loc[orden_partidos],
    fmt='',
    cmap="YlGnBu",
    linewidths=0.5,
    cbar_kws={'label': 'Percentile'}
)
plt.title("Categorical Heatmap by Party and Block")
plt.xlabel("Thematic blocks")
plt.ylabel("Political party")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("heatmap_categorico_bloques.png", dpi=300)
plt.show()

# 🔵 TERCER GRÁFICO: Evolución de la activación
plt.figure(figsize=(10,6))

for partido in orden_partidos:
    plt.plot(
        promedio_bloques_por_partido.columns,
        promedio_bloques_por_partido.loc[partido],
        marker='o',
        label=partido
    )

plt.title("Evolution of Average Activation by Thematic Block")
plt.xlabel("Thematic blocks")
plt.ylabel("Average activation")
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title="Political party", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("evolucion_activacion_bloques.png", dpi=300)
plt.show()

```

```python
import matplotlib.pyplot as plt

# 1. Orden de partidos y normalización
orden_partidos = ['Podemos', 'Sumar', 'PSOE', 'PP', 'Vox']
datos = promedio_bloques_por_partido.loc[orden_partidos]
datos_normalizados = datos.div(datos.sum(axis=1), axis=0)

# 2. Colores
bloques_temas = {
    'Anti-elite discourse': 'skyblue',
    'People': 'lightgreen',
    'Popular will': 'gold',
    'Left-wing adversaries': 'salmon',
    'Right-wing adversaries': 'plum'
}

# 3. Crear el gráfico
fig, ax = plt.subplots(figsize=(10, 6))

bottom = [0] * len(datos_normalizados)  # Inicializar la base para apilar

# Dibuja cada bloque
for bloque, color in bloques_temas.items():
    valores = datos_normalizados[bloque]
    bars = ax.bar(
        orden_partidos,
        valores,
        bottom=bottom,
        label=bloque,
        color=color
    )

    # Agregar etiquetas de porcentaje
    for bar, valor in zip(bars, valores):
        if valor > 0.02:  # Solo si la sección es suficientemente grande (>2%)
            ax.text(
                bar.get_x() + bar.get_width()/2,
                bar.get_y() + bar.get_height()/2,
                f'{valor*100:.1f}%',
                ha='center', va='center',
                fontsize=8, color='black'
            )
    
    # Actualizar el "bottom" para apilar
    bottom = [i + j for i, j in zip(bottom, valores)]

# 4. Formato
ax.set_ylabel('Proportion')
ax.set_title('Proportion of Topics by Political Party')
ax.legend(title='Block', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig("Proportion of Topics by Political Party", dpi=300)
plt.show()
```
